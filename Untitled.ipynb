```python
import numpy as np
```

```python
import os
```

```python
distance_file = os.path.join('data', 'distance_data_headers.csv')
```

```python
distances = np.genfromtxt(fname=distance_file, delimiter=',')
```

```python
print (distances)
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = distances[1:]
data = data.astype(np.float)
```

```python
plt.figure()     #This initializes a new figure
plt.plot(data[:,1])
```

```python
plt.figure()
plt.plot(data[:,1])
```

```python
headers = distances[0]
for col in range(1, len(headers)):
    fig = plt.plot(data[:,col], label=headers[col])
    plt.legend()

plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.savefig('all_samples.png')
```

```python
plt.figure()
plt.plot(data[0::100,0], data[0::100,1],'--or');
```
